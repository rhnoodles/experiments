```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Prepare sample data
sections = ['4 KB', '2 MB', '1 GB']
x_labels = ['HostToDevice A', 'HostToDevice B', 'DeviceToHost C']
data = {
    '4 KB': [2940588676, 2862997200, 5689044986],
    '2 MB': [1316361188, 1316311326, 5190804908],
    '1 GB': [1241911002, 1241057262, 2738976118]
}

# Step 2: Plot settings
fig, axes = plt.subplots(1, 3, figsize=(18, 6))  # Create 3 subplots for 3 sections

for i, section in enumerate(sections):
    # Bar plot for each section
    axes[i].bar(x_labels, data[section], color=['red', 'blue', 'green'])
    axes[i].set_title(section, fontsize=24)
    axes[i].set_ylim(0, 5789044986)
    axes[i].set_ylabel('Duration (ns)', fontsize=24)
    #axes[i].set_xlabel('Categories')
    axes[i].grid()

# Adjust layout
plt.tight_layout()
plt.show()
```
